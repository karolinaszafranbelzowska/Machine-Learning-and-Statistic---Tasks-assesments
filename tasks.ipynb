Square root of 2 is a positive real number whose square is equal to 2. So this is an example of an algebraic number of degree 2. Geometrically the square root of 2 is the length of the diagonal of a square with side length 1, which follows directly from the Pythagorean theorem[3].
Many methods of calculating the square root of a positive real number S require an initial value. If this value is too far from the actual value of the root, the calculation will be considerably longer. Accordingly, it is highly desirable to have an estimate of this amount, which may even be very inaccurate but simple to derive[1].

The Square Roots via Newton’s Method is a classic algorithm that illustrates many of concerns. The algorithm starts with some guess x1 > 0 and computes the sequence of improved guesses[4].


**Digit by digit determination** is a method of finding successive digits of the square root. It is slower than the Babylonian method, but has several advantages:
- it is simple for manual calculations;
- each digit of the determined root is correct, that is, it will not change in the course of further calculations;
- if the square root has a finite expansion, the algorithm ends after finding the last digit. This can be used to check whether a given integer is a square.

### The Newton's methon to calculate the square root of 2

```python
# A function sqrt2() to calculate the square root of 2 using Newton's method [5]

def sqrt2(number, number_iters = 500):
    
    a = float(number) # number to get square root of = a
    
    # a = 2
    
    for i in range(number_iters): # iteration number to meet desired accuracy.
        
        number = 0.5 * (number + a / number) # update
        
# x_(n+1) = 0.5 * (x_n +a / x_n)
    
    return number


print(sqrt2(2)) # calling sqrt2 (newton_method) which gives me the square root of 2
# Output: 1.41421356237
```

To display the square root of 2 to 100 decimal places I will use the proper format to get enough decimal places.

```python
b = sqrt2(2)
format(b, '.100f') # to get 100 digits after the decimal point I used '.100f' format[6]
```

The Chi-square test: x2
– any test statistic where the test statistic has a chi-square distribution if the theoretical relationship is true. The chi-square test is for testing hypotheses. In other words, the test value is judged by the chi-square distribution. The test most often used in practice. We can use it to test the compliance of both measurable and non-measurable features[8].
The Chi-Squared test is a statistical hypothesis test that assumes (the null hypothesis) that the observed frequencies for a categorical variable match the expected frequencies for the categorical variable. The test calculates a statistic that has a chi-squared distribution, named for the Greek capital letter Chi (X) pronounced “ki” as in kite. The result of the test is a test statistic that has a chi-squared distribution and can be interpreted to reject or fail to reject the assumption or null hypothesis that the observed and expected frequencies are the same[9].

**The Pearson’s chi-squared test** for independence can be calculated in Python using the _chi2_contingency() SciPy function_[10].
The Pearson’s Chi-Square statistical hypothesis is a test for independence between categorical variables. I will perform the test using Python’s SciPy module[11}.

```python
# import required libraries
from scipy.stats import chi2_contingency # this function computes the chi-square statistic and p-value 
                                         # for the hypothesis test of independence of the observed frequencies in 
                                         # the contingency table observed.   
from scipy.stats import chi2
import numpy as np # to make my code easier to read.
import pandas as pd # for dataframe work.
```

```python
table = np.array([[90, 60, 104, 95], # table helps to perform the test.
                 [30, 50, 51, 20], 
                 [30, 40, 45, 35]])
print(table)
```

```python
# Create df with table data.
data = {'A':[90, 30, 30, 150], 'B':[60, 50, 40, 150], 'C':[104, 51, 45, 200],
        'D':[95, 20, 35, 150], 'Total':[349, 151, 150, 650]}
df = pd.DataFrame(data,index=['White collar', 'Blue collar', 'No Collar', 'Total'])
df
```

```python
stat, p, dof, expected = chi2_contingency(table) # Perform the test. 
                                                 # returns 3 values, the Chi-square statistic (stat), p-value (p) 
                                                 # and degrees of freedom (dof) 

print(f"p-Value: {p}") # p-value
print('dof = %d' % dof ) #degrees of freedom
print(f"Chi-square stat: {stat}\n")

print(expected) # prints a table of expected values
```

```python
chi2_contingency(table)
```

```python
alpha = 0.05 # alpha value of 0.05[11]
print("p = " + str(p)) 
if p <= alpha: # interpret p-value 
    print('Dependent (reject H0)') 
else: 
    print('Independent (H0 holds true)') 
```

The **alpha** value denotes the probability of erroneously rejecting H0 (the null hypothesis) when it is true. A lower alpha value is chosen in cases where we expect more precision. If the p-value for the test comes out to be strictly greater than the alpha value, then H0 holds true.
Therefore, I rejet H0, that is, the result is statistically significant[12].

The standard deviation of an array of numbers x is calculated using numpy as np.sqrt(np.sum((x - np.mean(x))**2)/len(x)). However, Microsoft Excel has two different versions of the standard deviation calculation, STDEV.P and STDEV.S . The STDEV.P function performs the above calculation but in the STDEV.S calculation the division is by len(x)-1 rather than len(x) . Research these Excel functions, writing a note in a Markdown cell about the difference between them. Then use numpy to perform a simulation demonstrating that the STDEV.S calculation is a better estimate for the standard deviation of a population when performed on a sample. Note that part of this task is to figure out the terminology in the previous sentence.

***

**STDEV.P function**[14] - calculates standard deviation based on the entire population given as arguments (ignores logical values and text).
The standard deviation is a measure of how widely values are dispersed from the average value (the mean).
**STDEV.S function**[15] - estimates standard deviation based on a sample (ignores logical values and text in the sample).
The standard deviation is a measure of how widely values are dispersed from the average value (the mean).

Population and sample standard deviation
Standard deviation measures the spread of a data distribution. It measures the typical distance between each data point and the mean.
The formula we use for standard deviation depends on whether the data is being considered a population of its own, or the data is a sample representing a larger population[16].
- If the data is being considered a population on its own, we divide by the number of data points, NNN.
- If the data is a sample from a larger population, we divide by one fewer than the number of data points in the sample, n-1n−1n, minus, 1.

```python
# import required libraries
import numpy as np # numerical array
import random
import matplotlib.pyplot as plt # plotting
import statistics # describing and summarizing data
import pandas as pd # statistical summary of data frame.
```

```python
p = np.random.randint(1,500,50) # Create a population of random integers between 1 and 500, size = 50
p
```

```python
n = np.size(p)
n
```

```python
figure = plt.hist(p, bins=25)
plt.title("Population", fontsize = 20)
plt.show(n)
```

```python
# Printing standard deviation [17]
print("The standard deviation of population (p) is % s " 
                % (statistics.stdev(p))) 
# variance is approximately the squared result of what stdev is.
print("Variance of the population is % s" 
     %(statistics.variance(p))) 
```

```python
sd_p = np.std(p)
print("(STDEV.P) The standard deviation of the population is", sd_p)
```

```python
# Take a sample of 20 from the population 
p[0:20] 
sample = p[0:20]
sample
```

```python
figure = plt.hist(sample, bins=25)
plt.title("Sample of population", fontsize = 20)
plt.show()
```

```python
sd_sample = np.std(sample)
print("The standard deviation of sample of population is", sd_sample)
```

```python
# standard deviation of sample using a sample of data
s_sample = np.std(sample,ddof=1) # With this implementation, I can use ddof=0 to calculate the standard deviation 
                                  # of a population, or I can use ddof=1 to estimate the standard deviation 
                                  # of a population using a sample of data[18].
print("(STDEV.S) The standard deviation of a population using a sample of data", s_sample)
```

```python
df = pd.DataFrame(sample)
print (df.describe())
```

The plot shows the distribution of 20 numbers in the array. The describe() method confirms my earlier calculations.
Excel's STDEV calculates the sample standard deviation, while NumPy's std calculates the population standard deviation by default.
np.std(sample, ddof=1) calculates the standard deviation of s using the sample variance (i.e. dividing by n-1 rather than n.).
STDEV calculates an estimate of the population's standard deviation based on the sample, while STDEV.P calculates the sample's standard deviation, which is the population (if the sample is the whole population).

**k-means clustering** is a method of vector quantization, originally from signal processing, that aims to partition *n* observations into *k* clusters in which each observation belongs to the cluster with the nearest mean. 
The algorithm has a loose relationship to the k-nearest neighbor classifier, a popular machine learning technique for classification that is often confused with k-means due to the name. Applying the 1-nearest neighbor classifier to the cluster centers obtained by k-means classifies new data into the existing clusters. This is known as nearest centroid classifier or Rocchio algorithm.[22]
The k-means clustering method is an unsupervised machine learning technique used to identify clusters of data objects in a dataset. There are many different types of clustering methods, but k-means is one of the oldest and most approachable.[23]

```python
# instantiate and fit Kmeans cluster model. I am going to use 3 clusters and a random_state of 0.
# random_state controls the random number generator used and n_clusters observations from data for the initial centroids.
km = KMeans(n_clusters = 3, random_state=0).fit(X) # random initialization of centroids and fit
# km.fit(X)
```

```python
# find the center points of the data.
center = km.cluster_centers_
print(center)
```
